```python
import pandas as pd
import ast
import spacy
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('ingredients.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
temp_value = df['Ingredients'][0]
temp_value
```

```python
def list_to_text(temp_value):
    temp_value = ast.literal_eval(temp_value)
    temp_text = ' '.join(temp_value)
    doc = nlp(temp_text)
    filtered_tokens = [token.text for token in doc if not token.is_stop and not
                       token.is_punct and not token.like_num and token.text not in {',', '.'}]
    filtered_str = ' '.join(filtered_tokens)
    return filtered_str
```

```python
print(list_to_text(temp_value))
```

```python
df['Ingredients_text'] = df['Ingredients'].apply(list_to_text)
```

```python
recipes = df['Ingredients_text'].to_list() 
```

```python
recipes
```

```python
units = [
    'cup', 'tbsp', 'tsp', 'gms', 'ml', 'pound', 'pinch', 'large', 'extra', 'finely', 'optional', 'drained', 'shaken', 'mashed', 'chopped', 'taste', 'inch', 'special', 'equipment', 'seasoned'
]

# Function to extract core ingredients using spaCy and further filtering
def extract_ingredients(recipe):
    # Process the recipe string with spaCy
    doc = nlp(recipe)

    # List to hold core ingredients
    core_ingredients = []

    # Identify core ingredients using noun chunks from spaCy
    for chunk in doc.noun_chunks:
        ingredient = chunk.text.strip()
        # Filter out measurements and other non-ingredient words
        if not any(char.isdigit() for char in ingredient) and len(ingredient) > 1:
            core_ingredients.append(ingredient)

    # Further refine by removing units and descriptors
    refined_ingredients = []
    for ingredient in core_ingredients:
        tokens = re.split(r'\s+', ingredient)
        filtered_tokens = [token for token in tokens if token.lower() not in units]
        refined_ingredient = ' '.join(filtered_tokens)
        if refined_ingredient:
            refined_ingredients.append(refined_ingredient)

    return refined_ingredients
text_blob = ''
# Extract core ingredients for each recipe
for i, recipe in enumerate(recipes):
    core_ingredients = extract_ingredients(recipe)
    text_blob += ', '.join(core_ingredients)
print(text_blob)
```

```python
file_name = "output.txt"

# Write the string to the file
with open(file_name, "w", encoding="utf-8") as file:
    file.write(text_blob)
```

```python
def ingredients_list(temp):
    doc = nlp(temp)
    lemmatized_text = ' '.join([token.lemma_ for token in doc])
    new_value = lemmatized_text.lower().split()
    list1 = []
    for ingre in ingredients:
        if ingre.lower() in new_value:
            list1.append(ingre)
    return list1
    
```

```python
val = df['Ingredients_text'][0]
print(type(val))
print(ingredients_list(val))
```

```python
df['Ingredients_list'] = df['Ingredients_text'].apply(ingredients_list)
```

```python
ingredient_index = {ingredient: i for i, ingredient in enumerate(ingredients)}

def ingredients_to_vector(ingredients_list):
    # Initialize a zero vector
    vector = np.zeros(len(ingredients), dtype=int)
    
    # Set the corresponding indices to 1
    for ingredient in ingredients_list:
        if ingredient in ingredient_index:
            index = ingredient_index[ingredient]
            vector[index] = 1
    
    return vector
```

```python
df['Vector'] = df['Ingredients_list'].apply(ingredients_to_vector)
```

```python
df.drop('Vector', axis=1, inplace =True)
```

```python
def recommend_top_recipes(user_ingredients, df, top_n=5):
    """
    Recommend the top N most similar recipes based on user's list of ingredients.

    Args:
    user_ingredients (list of str): List of ingredients from the user's fridge or kitchen.
    df (pd.DataFrame): DataFrame containing recipes and their vectors.
    top_n (int): Number of top recipes to recommend.

    Returns:
    list of str: The names of the top N most similar recipes.
    """
    # Convert user ingredients list to vector
    user_vector = ingredients_to_vector(user_ingredients).reshape(1, -1)
    
    # Get all recipe vectors
    recipe_vectors = np.vstack(df['Vector'].values)
    
    # Compute cosine similarity between user's vector and recipe vectors
    similarities = cosine_similarity(user_vector, recipe_vectors).flatten()
    
    # Get indices of the top N most similar recipes
    top_indices = np.argsort(similarities)[-top_n:][::-1]
    
    # Return the names of the top N most similar recipes
    return df.iloc[top_indices]['Recipe ID'].tolist()
```

```python

user_ingredients = ['Butter', 'Sugar', 'Milk']
top_recipes = recommend_top_recipes(user_ingredients, df, top_n=5)
print('Top 5 most similar recipes are:')
for recipe in top_recipes:
    print(recipe)
```
